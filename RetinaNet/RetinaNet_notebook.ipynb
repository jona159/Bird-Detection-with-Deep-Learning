```python
from arcgis.learn import RetinaNet, prepare_data
import arcpy
```

```python
data_path = r'C:\Users\User\Desktop\barther_oie\beschrifteteObjekte'
```

```python
data = prepare_data(data_path, batch_size=2, chip_size=256, val_split_pct=0.2)
```

```python
data.show_batch(rows=8)
```

```python
model = RetinaNet(data)
```

```python
model.fit(20, checkpoint=True, early_stopping=True, tensorboard=True)
```

```python
model.show_results()
```

```python
model.save('Retinanet_Vogel')
```
